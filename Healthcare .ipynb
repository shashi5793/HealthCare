# Missing Values

Features like **Glucose, Blood Pressure, Skin thickness, Insulin, BMI** of a person can never be **Zero**. 
So the fact that our data set canotains Zero represents **missing values**.

```python
for i in cols:
    missing = 0
    missing = len(df[i][df[i] == 0])
    print(" "+i+ " : ", missing)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(121)
df["Glucose"].hist()
plt.subplot(122)
sns.distplot(df['Glucose'])
print( "Mean : ", df['Glucose'].mean() )
print("Mode : ",df['Glucose'].mode() )
print( "Median : ",  df['Glucose'].median() )
```

```python
plt.figure(figsize=(15,5))
plt.subplot(121)
df["BloodPressure"].hist()
plt.subplot(122)
sns.distplot(df["BloodPressure"])
print( "Mean : ", df['BloodPressure'].mean() )
print("Mode : ",df['BloodPressure'].mode() )
print( "Median : ",  df['BloodPressure'].median() )
```

```python
plt.figure(figsize=(15,5))
plt.subplot(121)
df["SkinThickness"].hist()
plt.subplot(122)
sns.distplot(df["SkinThickness"])
print( "Mean : ", df['SkinThickness'].mean() )
print("Mode : ",df['SkinThickness'].mode() )
print( "Median : ",  df['SkinThickness'].median() )
```

```python
plt.figure(figsize=(15,5))
plt.subplot(121)
df["BMI"].hist()
plt.subplot(122)
sns.distplot(df['BMI'])
print( "Mean : ", df['BMI'].mean() )
print("Mode : ",df['BMI'].mode() )
print( "Median : ",  df['BMI'].median() )
```

```python
for i in cols:
    df[i].replace(0, df[i].median(), inplace = True)
```

Looking at the above distributions we can conclude that the features have extreme values which can affect the model if we use **Mean** to replace our missing values.
Therefore, we will use **Median** in our missing values imputation.

```python
for i in cols:
    missing = 0
    missing = len(df[i][df[i] == 0])
    print(" "+i+ " : ", missing)
```

# Building Models and Checking Their Accuracies

```python
from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
X = df.drop('Outcome', axis = 1)
```

```python
Y = df['Outcome']
```

# Splitting data into Training and Testing

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, random_state = 42, test_size = 0.2)
```

# KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
```

```python
from sklearn import metrics
```

```python
k =  KNeighborsClassifier(n_neighbors = 7)
```

```python
k.fit(x_train,y_train)
```

```python
pred = k.predict(x_test)
```

```python
accuracy_score(pred, y_test)
```

```python
tp, fp, tn, fn = confusion_matrix(pred, y_test).ravel()
sensitivity = float(tp/(tp+fn))
specificity= float(tn/(tn+fp))
print("sensitivity:" ,float(tp/(tp+fn)))
print("specificity:", float(tn/(tn+fp)))
```

```python
tpr = sensitivity
fpr = 1-specificity 
```

```python
auc = roc_auc_score(y_test,pred)
auc
false_positive_rate, true_positive_rate, thresolds = metrics.roc_curve(y_test, pred)
display(false_positive_rate, true_positive_rate, thresolds)

plt.figure(figsize=(5, 8), dpi=100)
plt.axis('scaled')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.plot( false_positive_rate, true_positive_rate, 'g')
plt.fill_between(false_positive_rate, true_positive_rate, facecolor='lightgreen', alpha=0.7)
plt.text(0.95, 0.05, 'AUC = %0.4f' % auc, ha='right', fontsize=12, weight='bold', color='blue')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
```

# SVM Classifier

```python
from sklearn import svm
from sklearn import pipeline
from sklearn.preprocessing import StandardScaler
```

```python
clf = pipeline.make_pipeline(StandardScaler(), svm.LinearSVC(random_state = 0))
clf.fit(x_train, y_train)
```

```python
svm_pred = clf.predict(x_test)
```

```python
accuracy_score(y_test, svm_pred)
```

```python
confusion_matrix(y_test, svm_pred)
```

```python
tp, fp, tn, fn = confusion_matrix(y_test, svm_pred).ravel()
```

```python
print(" true positive:", tp)
print(" false positive:", fp)
print(" true negative:", tn)
print(" false negative:", fn)
```

```python
sensitivity = tp/(tp+fn)
specificity = tn/(tn+fp)
print("sensitivity:",sensitivity)
print("specificity:",specificity)
```

```python
auc = metrics.roc_auc_score(y_test, svm_pred)
fpr , tpr , th = metrics.roc_curve(y_test, svm_pred)
plt.figure(figsize = (10,8))
plt.axis('scaled')
plt.xlim([0,1])
plt.ylim([0,1])
plt.plot(fpr, tpr)
plt.fill_between(fpr, tpr, facecolor='lightgreen', alpha=0.7 )
plt.text(0.4,0.6,'AUC = %0.4f' % auc, ha='right', fontsize=12, weight='bold', color='blue')
plt.show()
```
